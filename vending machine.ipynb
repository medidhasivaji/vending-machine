```python
numeric_cols = [
    'quantity_refilled','quantity_sold','current_stock', 'unit_price','revenue_generated','temperature_C','avg_customer_rating','downtime_minutes', 'power_consumption_kwh',
    'network_latency_ms', 'maintenance_flag'] # List of numeric columns
print(df[numeric_cols])
```

```python
nan_counts = df[numeric_cols].isna().sum() # Check for NaN values in the numeric columns
print(nan_counts)#
```

```python
# Fill NaN values in numeric columns with 0
df[numeric_cols] = df[numeric_cols].fillna(0)
nan_counts = df[numeric_cols].isna().sum() 
print(nan_counts)  # Verify no NaN values remain
```

```python
duplicates = df.duplicated(numeric_cols)
print(duplicates.sum())  # Print number of duplicate rows based on numeric columns
```

```python
text_columns = [
    'transaction_id', 'machine_id', 'location_city', 'location_type', 'product_id', 'product_name',
    'refill_date', 'refill_time','product_category', 'refill_person_id', 'refill_person_name',
    'machine_status', 'payment_method', 'maintenance_flag'
]
print(df[text_columns])
```

```python
nan_counts = df[text_columns].isna().sum() # Check for NaN values in the text columns
print(nan_counts)#
```

```python
df[text_columns] = df[text_columns].fillna('Unknown') # Fill NaN values in text columns with 'Unknown'
nan_counts = df[text_columns].isna().sum()  
print(nan_counts)  # Verify no NaN values remain
```

```python
duplicates = df.duplicated(text_columns)
print(duplicates.sum())
```

```python
df.columns
```

```python
df['refill_time'] = df['refill_time'].replace('Unknown', '00:00:00').fillna('00:00:00') # Handle 'Unknown' and NaN in refill_time

df['refill_datetime'] = pd.to_datetime(
df['refill_date'].astype(str).str.strip() + ' ' + df['refill_time'].astype(str).str.strip(),
    errors='coerce',
    dayfirst=True) # Combine refill_date and refill_time into refill_datetime

print(df['refill_datetime'].head())



```

```python
# Handle negative quantity_sold values
df['quantity_sold'] = df['quantity_sold'].abs()
print(df['quantity_sold'])
```

```python
df.to_csv('modified_vending_refill.csv', index=False)  # `index=False` avoids saving the index as a column
```

```python
# Total Revenue ---
total_revenue = df["revenue_generated"].sum()
revenue_per_machine = df.groupby("machine_id")["revenue_generated"].sum().sort_values(ascending=False) # Revenue per machine
print(f"Total Revenue generated across all vending machines: ₹{total_revenue}")
print("Revenue generated per machine:")
print(revenue_per_machine)
 
```

```python
plt.figure(figsize=(10,5))
revenue_per_machine.head(5).plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title("Top 5 Total Revenue generated across all vending machines", fontsize=14, fontweight='bold')
plt.xlabel("Machine ID")
plt.ylabel("Total Revenue (₹)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by city and calculate total revenue
city_sales = (
    df.groupby("location_city")["revenue_generated"]
    .sum()
    .sort_values(ascending=False)
)

# Create a dictionary to store rounded values
city_sales_with_round_off = {}

for city, revenue in city_sales.items():
    if revenue >= 1_000_000:
        formatted = f"{revenue/1_000_000:.4f}M"
    else:
        formatted = f"{revenue/1_000:.4f}K"
    city_sales_with_round_off[city] = formatted

# Print in normal readable format
print("Revenue generated per city (rounded to 4 decimals):")
for city in city_sales_with_round_off:
    print(f"{city}: {city_sales_with_round_off[city]}")
```

```python
product_revenue = df.groupby("product_name")["revenue_generated"].sum().sort_values(ascending=False) # Revenue per product
print("Revenue generated per product:")
product_revenue_with_round_off = {}                # Create a dictionary to store rounded values                             
for product, revenue in product_revenue.items():
    if revenue >= 1_000_000:
        formatted = f"{revenue/1_000_000:.4f}M"
    else:
        formatted = f"{revenue/1_000:.4f}K"
    product_revenue_with_round_off[product] = formatted 
print("\nFormatted Revenue (rounded to 4 decimals) for Each Product:")
for product, formatted_revenue in product_revenue_with_round_off.items():
    print(f"{product}: {formatted_revenue}")
```

```python
top_product = product_revenue.head(10)  # Top 10 products by revenue

fig, ax = plt.subplots(figsize=(10, 6))
top_product.plot(kind='bar', ax=ax, color='mediumseagreen', edgecolor='black')
ax.set_title("Top 10 Products by Total Revenue", fontsize=16, fontweight='bold', color='darkgreen')
ax.set_xlabel("Product Name", fontsize=12)
ax.set_ylabel("Total Revenue (₹)", fontsize=12)
ax.set_xticklabels(top_product.index, rotation=45, ha='right')
ax.grid(axis='y', linestyle='--', alpha=0.6)
for i, p in enumerate(ax.patches):
    height = p.get_height()
   
    product_name = top_product.index[i]
    formatted_revenue = product_revenue_with_round_off.get(product_name, 'revenue')  
    ax.text(
        p.get_x() + p.get_width() / 2, 
        height + (max(top_product) * 0.02), 
        f'{formatted_revenue}',  
        ha='center',  
        va='bottom', 
        fontsize=10,
        color='black'
    )
plt.tight_layout()
plt.show()
```

```python
# City-wise Product Revenues

# Pivot data:rows = cities, columns = products, values = revenue
city_product_revenue = df.pivot_table(
    index='location_city',
    columns='product_name',
    values='revenue_generated',
    aggfunc='sum',
    fill_value=0
)

# cities by total revenue
city_total_revenue = city_product_revenue.sum(axis=1).sort_values(ascending=False)
city_product_revenue = city_product_revenue.loc[city_total_revenue.index]
print(city_product_revenue)
print(city_product_revenue)
```

```python
import matplotlib.colors as mcolors

base_colors = list(mcolors.TABLEAU_COLORS.values())  # light-friendly colors
colors = [mcolors.to_rgba(c, alpha=0.6) for c in base_colors]  # slight transparency

plt.figure(figsize=(12,7))
bottom = [0] * len(city_product_revenue)  # initial bottom positions

for i, product in enumerate(city_product_revenue.columns):
    revenues = city_product_revenue[product].values
    bars = plt.bar( city_product_revenue.index, revenues, bottom=bottom,color=colors[i % len(colors)], edgecolor='black',
        linewidth=0.5,
        label=product
    )
    
    # Add horizontal dotted line for each product division (full bar width)
    for j, rev in enumerate(revenues):
        if rev > 0:
            plt.hlines(
                y=bottom[j] + rev,         # top of current segment
                xmin=j - 0.4,              # left edge of bar
                xmax=j + 0.4,              # right edge of bar
                colors='black',
                linestyles='dotted',
                linewidth=0.8
            )
    
    # Add labels on each segment in white
    for j, rev in enumerate(revenues):
        if rev > 0:
            if rev >= 1_000_000:
                formatted = f"{rev/1_000_000:.2f}M"
            elif rev >= 1_000:
                formatted = f"{rev/1_000:.2f}K"
            else:
                formatted = f"{rev:.0f}"
            plt.text( j,bottom[j] + rev/2,formatted, ha='center',
                va='center',
                fontsize=7,
                color='white'
            )
    
    # Update bottom for next product
    bottom = [bottom[k] + revenues[k] for k in range(len(revenues))]

# Add total revenue at top of each bar in black
for j, total in enumerate(city_total_revenue):
    if total >= 1_000_000:
        formatted_total = f"{total/1_000_000:.2f}M"
    elif total >= 1_000:
        formatted_total = f"{total/1_000:.2f}K"
    else:
        formatted_total = f"{total:.0f}"
    plt.text(
        j,
        total + max(city_total_revenue)*0.01,  # slight offset above bar
        formatted_total,ha='center',va='bottom',fontsize=8,
        fontweight='bold',
        color='black'
    )

plt.title("City-wise Product Revenue", fontsize=12, weight='bold', color='navy')
plt.xlabel("City", fontsize=12)
plt.ylabel("Revenue", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(title="Products", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df["month"] = df["refill_datetime"].dt.to_period("M") # 
monthly_revenue = df.groupby("month")["revenue_generated"].sum().sort_index()
```

```python
plt.figure(figsize=(10,5))
plt.plot(monthly_revenue.index.astype(str), monthly_revenue.values, marker='o', color='teal')
plt.title(" Monthly Revenue Trend", fontsize=14, fontweight='bold')
plt.xlabel("Month")
plt.ylabel("Total Revenue (₹)")
plt.grid(True)
plt.show()
```

```python
top_5_downtime = df.nlargest(5, "downtime_minutes")
top_5_downtime.set_index("machine_id")[["downtime_minutes", "revenue_generated"]].plot(kind='bar', figsize=(8, 4))
plt.title("Low 5 Machines: Downtime vs Revenue")
plt.xlabel("Machine ID")
plt.ylabel("Value")
plt.show()
```

```python

downtime = df[df["machine_status"].isin(["Low Power", "Maintenance"])]

# Count downtime events per machine
downtime_count = downtime.groupby("machine_id")["machine_status"].count().reset_index(name="downtime_count")

machine_performance = pd.merge(revenue_per_machine, downtime_count, on="machine_id", how="left").fillna(0)

top_5_downtime = machine_performance.nlargest(5, "downtime_count")
print(top_5_downtime.head())

fig, ax = plt.subplots(figsize=(10, 6))
top_5_downtime[["downtime_count"]].plot(kind='bar', ax=ax, color=["teal", "orange"], width=0.8)

ax.set_title("Top 5 Machines with Most Downtime ", fontsize=14, fontweight="bold")
ax.set_xlabel("Machine ID")
ax.set_ylabel("Count / Revenue")
ax.set_xticklabels(top_5_downtime["machine_id"], rotation=45, ha="right")

ax.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```
